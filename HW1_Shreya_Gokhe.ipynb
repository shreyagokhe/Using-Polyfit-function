##### 1.Plotting a polynomial y= 5*x + 20*x2 + x3 with the noisy data on the given dataset with m = 4

```python
plt.style.use('seaborn-whitegrid')
import numpy as np
noise_scale = 100
number_of_samples = 50
x = 25*(np.random.rand(number_of_samples, 1) - 0.8)
y = 5 * x + 20 * x**2 + 1 * x**3 + noise_scale*np.random.randn(number_of_samples, 1)
plt.plot(x,y,'ro')
graph_line = np.polyfit(x.flatten(),y.flatten(),4)
a= np.poly1d(graph_line)
x1= np.linspace(-20,5,50)
plt.plot(x.flatten(),y.flatten(),'ro',x1,a(x1),'b--')
```

### Conclusion: We get a sinusoidal graph with the plotted data points

MSE vs m where m=1

```python
def polyfun(noise_scale,number_of_samples,m):
    plt.style.use('seaborn-whitegrid')
    plt.title('Plotting MSE vs m')
    plt.xlabel(r'm')
    plt.ylabel(r'MSE')
    #noise_scale = 100
    #number_of_samples = 50
    x = 25*(np.random.rand(number_of_samples, 1) - 0.8)
    y = 5 * x + 20 * x**2 + 1 * x**3 + noise_scale*np.random.randn(number_of_samples,1)
    plt.plot(x,y,'ro')
    graph_line = np.polyfit(x.flatten(),y.flatten(),m)
    a= np.poly1d(graph_line)
    x1= np.linspace(-20,5,50)
    plt.plot(x.flatten(),y.flatten(),'ro',x1,a(x1),'b--')
    y_pred = a(x)
    mse = np.sum((y_pred-y)**2)/50
    print(mse)
    return mse
```

```python
mse_1=polyfun(100,50,1)
```
